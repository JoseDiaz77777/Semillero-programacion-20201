### 4. NLP

```python
nlp=spacy.load('es_core_news_sm')
doc=nlp(Titulares)
doc2=nlp(Contenido)
```

```python
type(doc)
```

```python
for element in doc:
    print(element)
```

```python
print(doc[12].text)
print(doc[12].pos_)
print(doc[12].lemma_)
```

#### Tokenizar:


```python
oraciones={}
for num,oracion in enumerate(doc2.sents):
    oraciones[num]= oracion
```

```python
oraciones
```

```python
oraciones_buscadas=[]
for oracion in oraciones.values():
    if "farc"in oracion.text.lower():
        oraciones_buscadas.append(oracion.orth_)
print(len(oraciones_buscadas))
print(oraciones_buscadas[0:5])
type(oraciones_buscadas[0])
```

```python
oraciones=",".join(oraciones_buscadas)
doc3=nlp(oraciones)
```

```python
#Pos-tag
palabras=""
pos_tag_target=["ADJ"]
for palabra in doc3:
    if palabra.text not in string.punctuation :
        if palabra.pos_ in pos_tag_target:
            palabras=palabras +" "+palabra.text.lower()
palabras
```

```python
#Hacer una nube de palabras de adjetivos

wordcloud = WordCloud(background_color ='white',stopwords=stopwords_es,collocations=False,width=1600, height=800).generate(palabras) 
  
# plot the WordCloud image                        
plt.figure( figsize=(20,10)) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.savefig('./Verbos_OracionesFarc.png')
plt.show() 
```
